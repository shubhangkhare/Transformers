```python
from transformers import pipeline

# Load the summarization pipeline
summarization_pipeline = pipeline("summarization")
```

```python
# Text to be summarized
text = '''ndia, officially known as the Republic of India, is a vibrant and diverse country located in South Asia. With a rich cultural heritage, historical significance, and a population of over 1.3 billion people, India is the second-most populous country in the world. It spans a vast land area, covering approximately 3.29 million square kilometers.

India is known for its incredible diversity, which is reflected in its languages, religions, cuisines, traditions, and landscapes. The country is home to numerous ethnic groups, each with its own distinct customs and practices. Hindi is the official language of the Indian government, but there are also 21 other officially recognized languages, including Bengali, Telugu, Marathi, Tamil, Urdu, Gujarati, and Punjabi, among others. English, inherited from the colonial era, serves as a subsidiary official language and is widely spoken.

India's history dates back thousands of years, with ancient civilizations such as the Indus Valley Civilization flourishing on its soil. The country has witnessed the rise and fall of powerful empires, including the Maurya, Gupta, Mughal, and Maratha empires, which have left a lasting impact on its cultural and architectural heritage. The influence of Buddhism, Jainism, Hinduism, and Islam can be seen in the numerous temples, mosques, stupas, and monuments that dot the Indian landscape.

India gained independence from British colonial rule on August 15, 1947, under the leadership of Mahatma Gandhi and other prominent freedom fighters. It adopted a democratic system of governance and established itself as a sovereign nation. The Indian constitution, implemented in 1950, enshrines principles of equality, liberty, and justice for all citizens.

The Indian economy is one of the world's largest by nominal GDP and is classified as a developing economy. It encompasses diverse sectors such as agriculture, manufacturing, services, information technology, pharmaceuticals, and textiles. India is also known for its vibrant film industry, popularly referred to as Bollywood, which produces the largest number of films globally.

India's cultural tapestry is woven with a myriad of festivals celebrated throughout the year. Diwali, Holi, Eid, Christmas, Navaratri, and Durga Puja are just a few examples of the festivals that bring people together in celebration and create an atmosphere of joy and unity. These festivals showcase the country's cultural richness and provide a glimpse into its traditions and customs.

In terms of natural beauty, India offers a wide range of landscapes, from the snow-capped Himalayas in the north to the pristine beaches of Goa and Kerala in the south. The country is also blessed with diverse ecosystems, including lush forests, expansive deserts, fertile plains, and serene backwaters. Wildlife enthusiasts can explore India's national parks and reserves, which are home to iconic species such as tigers, elephants, lions, rhinoceroses, and various bird species.

Indian cuisine is renowned worldwide for its flavors, spices, and regional specialties. From the aromatic biryanis of Hyderabad to the street food delights of Mumbai and the dosas of South India, the country offers a gastronomic adventure for every palate. Indian cuisine is characterized by its use of herbs, spices, and diverse cooking techniques that vary from region to region.

Despite its many achievements, India also faces challenges such as poverty, inequality, healthcare access, and environmental sustainability. The government and various organizations are actively working towards addressing these issues and fostering inclusive growth and development.

In conclusion, India is a country that captivates with its vibrant culture, rich history, diverse traditions, and stunning landscapes. It is a nation where modernity and tradition coexist, where ancient temples stand alongside bustling metropolises, and where the warmth and hospitality of its people leave a lasting impression'''
```

```python
# Generate the summary
summary = summarization_pipeline(text, max_length=150, min_length=30, do_sample=False)

# Print the summary
print(summary[0]['summary_text'])
```
