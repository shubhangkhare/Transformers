```python
from transformers import pipeline

pipe = pipeline("token-classification", model="dbmdz/bert-large-cased-finetuned-conll03-english")
```

```python
pipe('My name is Wolfgang and I live in Berlin')
```

```python
pipe.save_pretrained('/content/bert/')
```

```python
def split_file(input_file, output_file1, output_file2):
    # Read the input file
    with open(input_file, 'rb') as f:
        data = f.read()

    # Determine the split point
    midpoint = len(data) // 2

    # Write the first half to output_file1
    with open(output_file1, 'wb') as f1:
        f1.write(data[:midpoint])

    # Write the second half to output_file2
    with open(output_file2, 'wb') as f2:
        f2.write(data[midpoint:])

# Usage
split_file('/content/bert/model.safetensors', 'sample_data/part1.safetensors', 'sample_data/part2.safetensors')
```

```python
def join_files(input_file1, input_file2, output_file):
    # Read the first input file
    with open(input_file1, 'rb') as f1:
        data1 = f1.read()

    # Read the second input file
    with open(input_file2, 'rb') as f2:
        data2 = f2.read()

    # Write the combined data to the output file
    with open(output_file, 'wb') as f:
        f.write(data1)
        f.write(data2)

# Usage
join_files('sample_data/part1.safetensors', 'sample_data/part2.safetensors', '/content/bert/model.safetensors')
```
