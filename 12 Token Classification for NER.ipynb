# Load Model

```python
# Use a pipeline as a high-level helper
from transformers import pipeline
```

```python
pipe1 = pipeline("token-classification", model="dslim/bert-base-NER", aggregation_strategy="simple")
```

```python
pipe2 = pipeline("token-classification", model="dslim/bert-base-NER-uncased", aggregation_strategy="simple")
```

```python
pipe3 = pipeline("token-classification", model="dslim/bert-large-NER", aggregation_strategy="simple")
```

# Inference

```python
def convert_subword_to_word(predictions):
    word_predictions = []
    current_word = ""

    for token in predictions:
        if token['word'].startswith('##'):
            current_word += token['word'][2:]
        else:
            if current_word:
                word_predictions[-1]['word'] += current_word
                current_word = ""
            word_predictions.append({
                'entity_group': token['entity_group'],
                'score': token['score'],
                'word': token['word'],
                'start': token['start'],
                'end': token['end']
            })

    if current_word:
        word_predictions[-1]['word'] += current_word

    return word_predictions
```

```python
predictions = pipe1('I am Shubhang Khare, and I am from lucknow UP India')
convert_subword_to_word(predictions)
```

```python
predictions = pipe2('I am Shubhang Khare, and I am from lucknow UP India')
convert_subword_to_word(predictions)
```

```python
predictions = pipe3('I am Shubhang Khare, and I am from lucknow UP India')
convert_subword_to_word(predictions)
```
